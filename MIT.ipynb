# The Massachusetts Institute of Technology (MIT)

**Shannon Gallant Ansari, GitHub username: ShannonGA**

The Massachusetts Institute of Technology (MIT) is a private university in Cambridge, Massachusetts. Established in 1861 by a graduate of the Rensselaer Poyltechinc Institute (my alma mater) who felt that Massachusetts should also have a technical college, MIT has since then been at the center of the development of modern technology and science. Since it was founded, MIT has been ranked among the top academic institutions in the world.
